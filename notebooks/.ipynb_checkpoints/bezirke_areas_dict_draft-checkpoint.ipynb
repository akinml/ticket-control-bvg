```python
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib
```

```python
# get the response in the form of html
wikiurl_sbahns="https://de.wikipedia.org/wiki/Verwaltungsgliederung_Berlins"
table_class="wikitable sortable zebra jquery-tablesorter"
response=requests.get(wikiurl_sbahns)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')

indiatable=soup.find_all('table', {'class':"wikitable"})
```

```python
indiatable[1]
```

```python
df_sbahns=pd.read_html(str(indiatable[1]))
# convert list to dataframe
df_sbahns=pd.DataFrame(df_sbahns[0])
print(df_sbahns.head())
```

```python
len(df_sbahns["Ortsteil"])
```

```python
bezirke = {}
```

```python
bezirke = df_sbahns.groupby(["Bezirk"])["Ortsteil"].agg(lambda x: list(x)).to_dict()
```
